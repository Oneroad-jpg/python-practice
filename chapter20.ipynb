```python
# 1. 道具箱から「requests」と「BeautifulSoup」を取り出す。
import requests
from bs4 import BeautifulSoup

# 2. 情報を取得したいWebページのアドレス（URL)を指定する
# (今回は、この目的のために用意されたシンプルな練習サイトを使います。)
url = 'http://example.com/'

# 3. 「requests」を使っって、指定したURLからWebページのデータを取得する
# （「運び屋さん」にデータを持ってきてもらうイメージです）
response = requests.get(url)

# 4. 「BeautifulSoup」を使って、取得したデータを解析（パース）する
# (料理人が読めるように、生の素材を整えるイメージです)
# respons.textで、取得したデータの中からHTMLのテキスト部分だけを取り出しています
soup = BeautifulSoup(response.text, 'html.parser')

# 5. 解析したデータから、ページの「タイトル(<title>タグ)を取り出して表示する
print(soup.title)

# 6. 解析したデータから、ページの「一番大きな見出し(<h1>タグ)を取り出して表示する
print(soup.h1)
```
