```python
# --- 1. バブルソート アルゴリズムの定義 ---
print("--- Defining Bubble Sort Algorithm ---")

def bubble_sort(list_data):
    # リストの長さを取得
    n = len(list_data)

    # B試験の理論：バブルソートは「二重ループ」が特徴

    # 外側のループ（パス。 n-1 回繰り返す）
    for i in range(n - 1):
        print(f"\n--- パス {i + 1} ---")

        # 内側のループ（隣同士の比較と交換）
        # (i)回ループが終わるたびに、リストの末尾(i)個はソート済みになる
        for j in range(n - i - 1):

            # 1. 隣同士 (j と　j+1) を比較
            if list_data[j] > list_data[j +1]:
                # 2. もし左(j)が右(j+1)より大きければ、交換(swap)する
                print(f" 交換：{list_data[j]} と {list_data[j + 1]}")
                list_data[j], list_data[j + 1] = list_data[j + 1], list_data[j] # Python流の交換
            else:
                print(f" 比較：{list_data[j]} と {list_data[j + 1]}")

    return list_data

# --- 2. バブルソート アルゴリズムを実行する ---
print("\n--- Executing Bubble Sort ---")

# ソートの対象のリスト
my_list = [6, 3, 8, 5, 2]

print(f"ソート前のリスト： {my_list}")

#  バブルソートの実行
sorted_list = bubble_sort(my_list)

print(f"\nソート後のリスト： {sorted_list}")
```
