```python
# --- 1. マージソート アルゴリズムの定義 ---
print("--- Defining Merge Sort Algorithm ---")

def merge_sort(list_data):
    # [再帰の停止条件]　リストの要素が１つ以下なら、ソート不要
    if len(list_data) <= 1:
        return list_data
    
    # 1. 分割 (Divide)
    # リストの真ん中を見つける
    mid = len(list_data) // 2

    # 真ん中を境に、左半分と右半分に分割
    left_half = list_data[:mid]
    right_half = list_data[mid:]

    print(f" 分割中：{list_data} -> 左：{left_half}, 右：{right_half}")

    # [再帰呼び出し]
    left_half_sorted = merge_sort(left_half)
    right_half_sorted = merge_sort(right_half)

    merged_list = []
    i = 0 # 左半分用のインデックス
    j = 0 # 右半分用のインデックス

    # 左(i)と右(j)の先頭を比較し、小さい方から merged_list に追加
    while i < len(left_half_sorted) and j < len(right_half_sorted):
        if left_half_sorted[i] < right_half_sorted[j]:
            merged_list.append(left_half_sorted[i])
            i += 1

        else:
            merged_list.append(right_half_sorted[j])
            j += 1
    
    # 比較が終わった後、残った要素を追加（iかjのどちらかがリストの末尾に達している）
    merged_list.extend(left_half_sorted[i:])
    merged_list.extend(right_half_sorted[j:])
    

    print(f" 併合完了：{merged_list}")
    return merged_list

# --- 2. マージソート アルゴリズムを実行する ---
print("\n--- Executing Merge Sort ---")

# ソートの対象リスト（バブルソートと同じ）
my_list = [6, 3, 8, 5, 2]

print(f"ソート前のリスト： {my_list}")

# マージソートを実行
sorted_list = merge_sort(my_list)

print(f"\nソート後のリスト：{sorted_list}")
```
